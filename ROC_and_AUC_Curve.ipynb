# Now, let's create the dummy data and plot the ROC curve:
This code generates a synthetic dataset, trains a logistic regression classifier, and plots the ROC curve with the area under the curve (AUC) displayed in the plot's label. The plot will show the trade-off between TPR (TPR) and FPR (x-axis) for different threshold settings.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc
from sklearn.linear_model import LogisticRegression

# Synthetic dataset
X = np.random.rand(1000, 2)
y = (X[:, 0] > X[:, 1]).astype(int)

# Train a logistic regression classifier
clf = LogisticRegression()
clf.fit(X, y)

# Predict probabilities
y_pred_proba = clf.predict_proba(X)[:, 1]

# Compute true positive rate (TPR) and false positive rate (FPR)
fpr, tpr, thresholds = roc_curve(y, y_pred_proba)

# Calculate AUC
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.plot(fpr, tpr, label='ROC curve (AUC = %.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], linestyle='--', color='black', label='Chance')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc='lower right')
plt.show()
```

# Now, let's create the dummy data and calculate the AUC:

This code generates a synthetic dataset, trains a logistic regression classifier, and calculates the AUC using the roc_curve and auc functions from sklearn.metrics. The calculated AUC value will be displayed in the console.

It's worth noting that you don't need to visualize the ROC curve to calculate the AUC. The AUC value summarizes the classifier's performance across all possible threshold settings.

```python
import numpy as np
import pandas as pd
from sklearn.metrics import roc_curve, auc
from sklearn.linear_model import LogisticRegression

# Synthetic dataset
X = np.random.rand(1000, 2)
y = (X[:, 0] > X[:, 1]).astype(int)

# Train a logistic regression classifier
clf = LogisticRegression()
clf.fit(X, y)

# Predict probabilities
y_pred_proba = clf.predict_proba(X)[:, 1]

# Compute true positive rate (TPR) and false positive rate (FPR)
fpr, tpr, thresholds = roc_curve(y, y_pred_proba)

# Calculate AUC
roc_auc = auc(fpr, tpr)

print("AUC:", roc_auc)
```
